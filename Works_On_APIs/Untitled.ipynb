```python
import requests
from bs4 import BeautifulSoup
from readability import Document

# URL of the News API endpoint
api_url = 'https://newsapi.org/v2/everything?q=Apple&sortBy=publishedAt&apiKey=97df821f255d46018574e4f2af172df2'

# Make the request to the News API
response = requests.get(api_url)
if response.status_code == 200:
    # Extract the URL of the first article from the API response
    first_article_url = response.json()['articles'][0]['url']
    
    # Make a request to fetch the HTML content of the first article
    article_response = requests.get(first_article_url)
    if article_response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(article_response.content, 'html.parser')
        
        # Use Readability to parse the article content
        document = Document(soup.prettify())
        article_content = document.summary()
        
        # Print the parsed article content
        print(article_content)
    else:
        print(f"Failed to fetch article HTML. Status code: {article_response.status_code}")
else:
    print(f"Failed to fetch News API data. Status code: {response.status_code}")
```
